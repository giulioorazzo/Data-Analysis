```python
high_GME =  GME['High GME']
GME['log_GME'] = np.log(high_GME)
# calculating the percentage change
GME['pctChangeGME'] = np.exp(GME['log_GME']).pct_change() * 100
```

```python
high_TNX =  TNX['High TNX']
TNX['log_TNX'] = np.log(high_TNX)
# calculating the percentage change
TNX['pctChangeTNX'] = np.exp(TNX['log_TNX']).pct_change() * 100
```

```python
high_THCX =  THCX['High THCX']
THCX['log_THCX'] = np.log(high_THCX)
# calculating the percentage change
THCX['pctChangeTHCX'] = np.exp(THCX['log_THCX']).pct_change() * 100
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df = pd.DataFrame({
    'pctChangeTHCX': THCX['pctChangeTHCX'],
    'pctChangeTNX': TNX['pctChangeTNX'],
    'pctChangeGME': GME['pctChangeGME']
})
# Drop rows with any NaN values
df = df.dropna()

X = df[['pctChangeTHCX', 'pctChangeTNX']]
y = df['pctChangeGME']
model = LinearRegression()

model.fit(X , y)

# Make predictions on the test data
y_pred = model.predict(X)

# Evaluate the model
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error: {mse}')

# Coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

print(f'\nCoefficients: {coefficients}')
print(f'Intercept: {intercept}')
```
